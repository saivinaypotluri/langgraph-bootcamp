### Prompts in Langgraph

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os

load_dotenv()

if os.environ['GOOGLE_API_KEY']:
    print("Google API key is set")
else:
    raise ValueError("Google API key is not set")
```

```python
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=1.0,  # Gemini 3.0+ defaults to 1.0
    max_tokens=None,
    timeout=None,
    max_retries=2,
   # other params...
)

try :
  print("Dont waste Tokens")
 # print(model.invoke("I want to know about Langgraph ?"))
except:
  print("Something went wrong")
```

```python
from langchain_core.prompts import ChatPromptTemplate

my_prompt=ChatPromptTemplate.from_messages(# It accepts a List
    [
        ("system","You are a helpful assistant that helps to create and curate posts for a social media platform."),
        ("human","I want to create a post about the {topic}")
    ]
)
```

```python
prompt_filled = my_prompt.invoke({"topic":"The importance of data privacy in the digital age"})
```

```python
model.invoke(prompt_filled)
```

### Chain with Prompts


```python
chain = my_prompt | model # This is a chain in Langchain, where output of my_prompt will go as input to model(llm)
```

```python
chain.invoke({"topic":"The importance of data privacy in the digital age"}) # This invoke acts as input to my_prompt(prompt), which generates output that acts as input to model(llm). The output which we are seeing is output of model(llm).
```
